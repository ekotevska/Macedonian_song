```
from note_seq.protobuf import music_pb2

po_tebe = music_pb2.NoteSequence()

# Add the notes to the sequence.
po_tebe.notes.add(pitch=68, start_time=0.0, end_time=0.5, velocity=80)
po_tebe.notes.add(pitch=70, start_time=0.5, end_time=1.0, velocity=80)
po_tebe.notes.add(pitch=68, start_time=1.0, end_time=1.5, velocity=80)
po_tebe.notes.add(pitch=66, start_time=1.5, end_time=2.0, velocity=80)
po_tebe.notes.add(pitch=70, start_time=2.0, end_time=2.5, velocity=80)
po_tebe.notes.add(pitch=70, start_time=2.7, end_time=3.2, velocity=80)
po_tebe.notes.add(pitch=68, start_time=3.4, end_time=3.9, velocity=80)
po_tebe.notes.add(pitch=70, start_time=3.9, end_time=4.4, velocity=80)
po_tebe.notes.add(pitch=68, start_time=4.4, end_time=4.9, velocity=80)
po_tebe.notes.add(pitch=65, start_time=4.9, end_time=5.4, velocity=80)
po_tebe.notes.add(pitch=63, start_time=5.4, end_time=5.9, velocity=80)

#pitch is the quality that makes it possible to judge sounds as "higher" and "lower"
#velocity is the force with which a note is played


po_tebe.total_time = 8 #the total time of the Sequence in seconds

po_tebe.tempos.add(qpm=60); #tempo is measured in quarter-notes per minute (or qpm)

# This is a colab utility method that visualizes a NoteSequence.
note_seq.plot_sequence(po_tebe)

# This is a colab utility method that plays a NoteSequence.
note_seq.play_sequence(po_tebe,synth=note_seq.fluidsynth)

```

```
ledena = music_pb2.NoteSequence()

# Add the notes to the sequence.
ledena.notes.add(pitch=64, start_time=0.0, end_time=0.4, velocity=80)
ledena.notes.add(pitch=64, start_time=0.4, end_time=0.8, velocity=80)
ledena.notes.add(pitch=63, start_time=0.8, end_time=1.2, velocity=80)
ledena.notes.add(pitch=64, start_time=1.2, end_time=1.6, velocity=80)
ledena.notes.add(pitch=66, start_time=1.6, end_time=2.0, velocity=80)
ledena.notes.add(pitch=64, start_time=2.0, end_time=2.4, velocity=80)
ledena.notes.add(pitch=63, start_time=2.4, end_time=2.8, velocity=80)
ledena.notes.add(pitch=64, start_time=2.8, end_time=3.2, velocity=80)
ledena.notes.add(pitch=66, start_time=3.2, end_time=3.6, velocity=80)
ledena.notes.add(pitch=64, start_time=3.6, end_time=4.0, velocity=80)
ledena.notes.add(pitch=63, start_time=4.0, end_time=4.4, velocity=80)
ledena.notes.add(pitch=64, start_time=4.4, end_time=4.8, velocity=80)
ledena.notes.add(pitch=68, start_time=4.8, end_time=5.2, velocity=80)
ledena.notes.add(pitch=66, start_time=5.2, end_time=5.6, velocity=80) 
ledena.notes.add(pitch=64, start_time=5.6, end_time=6.0, velocity=80)
ledena.notes.add(pitch=63, start_time=6.0, end_time=6.4, velocity=80)
ledena.total_time = 8

ledena.tempos.add(qpm=60);

# This is a colab utility method that visualizes a NoteSequence.
note_seq.plot_sequence(ledena)

# This is a colab utility method that plays a NoteSequence.
note_seq.play_sequence(ledena,synth=note_seq.fluidsynth)
```

**–ò–Ω–∏—Ü–∏—ò–∞–ª–∏–∑–∏—Ä–∞—ö–µ –Ω–∞ –º–æ–¥–µ–ª–æ—Ç Melody RNN**

```
print('Downloading model bundle. This will take less than a minute...')
note_seq.notebook_utils.download_bundle('basic_rnn.mag', '/content/')

# Import dependencies.
from magenta.models.melody_rnn import melody_rnn_sequence_generator
from magenta.models.shared import sequence_generator_bundle
from note_seq.protobuf import generator_pb2
from note_seq.protobuf import music_pb2

# Initialize the model.
print("Initializing Melody RNN...")
bundle = sequence_generator_bundle.read_bundle_file('/content/basic_rnn.mag')
generator_map = melody_rnn_sequence_generator.get_generator_map()
melody_rnn = generator_map['basic_rnn'](checkpoint=None, bundle=bundle)
melody_rnn.initialize()

print('üéâ Done!')
```

**–ü—Ä–æ–¥–æ–ª–∂—É–≤–∞—ö–µ –Ω–∞ –ø—Ä–≤–∞—Ç–∞ —Å–µ–∫–≤–µ–Ω—Ü–∞**

```
# Model options.  
input_sequence = po_tebe # the sequence that will be used
num_steps = 128 # num_steps is how long each melody will be in 16th steps (128 steps = 8 bars)
temperature = 1.5 # the higher the temperature the more random the sequence.

# Set the start time to begin on the next step after the last note ends.
last_end_time = (max(n.end_time for n in input_sequence.notes)
                  if input_sequence.notes else 0)
qpm = input_sequence.tempos[0].qpm #the tempo that is set before
seconds_per_step = 60.0 / qpm / melody_rnn.steps_per_quarter 
total_seconds = num_steps * seconds_per_step #equals end_time

generator_options = generator_pb2.GeneratorOptions()
generator_options.args['temperature'].float_value = temperature
generate_section = generator_options.generate_sections.add(
  start_time=last_end_time + seconds_per_step,
  end_time=total_seconds)

# Ask the model to continue the sequence.
sequence1 = melody_rnn.generate(input_sequence, generator_options)

note_seq.plot_sequence(sequence1)
note_seq.play_sequence(sequence1, synth=note_seq.fluidsynth)
```

```
note_seq.sequence_proto_to_midi_file(sequence1, 'po_tebe.mid')
files.download('po_tebe.mid')
```
